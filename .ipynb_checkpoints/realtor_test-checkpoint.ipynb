Below, I read in the csv files and answer your question

```python
df = pd.read_csv('zip_metro_mapping.csv',encoding='ISO-8859-1')
df1 = pd.read_csv('RDC_Inventory_Core_Metrics_Metro.csv')
df2 = pd.read_csv('RDC_Inventory_Core_Metrics_Zip.csv')
```

```python
df2.columns
```

```python
#### Removes unnecessary rows ####
df2 = df2.filter(items=['postal_code','zip_name','median_listing_price','active_listing_count'])
df2
```

```python
#### This line of code removes all entries where the median_listing_price is below $1 million
df_median_listing_price = df2[df2['median_listing_price'] > 999999 ]
```

```python
#### This line of code sums the 'active_listing_count' 
df_median_listing_price.active_listing_count.sum()
```

## The 20 most expensive zip codes in Pittsburgh, Pa

```python
df2_pittsburgh = df2[df2['zip_name'] == 'pittsburgh, pa'].sort_values(by='median_listing_price',\
ascending=False).head(20)

df2_pittsburgh
```

## The 20 most expensive zip codes in Philadelphia, Pa

```python
df2_philly = df2[df2['zip_name'] == 'philadelphia, pa'].sort_values(by='median_listing_price',\
ascending=False).head(20)

df2_philly
```

## The two joined together

```python
concatenated = pd.concat([df2_pittsburgh,df2_philly])
concatenated.sort_values('median_listing_price')
```
